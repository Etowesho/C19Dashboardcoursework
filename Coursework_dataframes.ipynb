```python
from uk_covid19 import Cov19API
import json
```

```python
filters=[
    'areaType=nation',
    'areaName=England'
]
structure={
    'date':'date',
    'newcases':'newCasesByPublishDate',
    'cases': 'cumCasesByPublishDate',
    'deaths':'cumDeathsByPublishDate'  
}
api=Cov19API(filters=filters, structure=structure)
data1=api.get_json()
print(data1) #date, all cases, new cases on that date and all deaths to date.
```

```python
filters=[
    'areaType=nation',
    'areaName=England'
]
structure={
    'date':'date',
    'admissionsage':'cumAdmissionsByAge',
    'addmissions':'cumAdmissions'
}
api=Cov19API(filters=filters, structure=structure)
data2=api.get_json()
print(data2) #new admissions and admissions by age
```

```python
filters=[
    'areaType=nation',
    'areaName=England'
]
structure={
    'date':'date',
    'newadmissions':'newAdmissions',
    'occupiedbeds':'covidOccupiedMVBeds'
}
api=Cov19API(filters=filters, structure=structure)
data3=api.get_json()
print(data3) #new admissions and occupied beds
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
%matplotlib inline

plt.rcParams['figure.dpi'] = 100
```

```python
with open("data1", "rt") as INFILE:
    Graphic1=json.load(INFILE)
```

```python
Graphic1 #newcases, cases, deaths
```

```python
datalist=Graphic1['data']
datalist
```

```python
dates=[dictionary['date'] for dictionary in datalist ]
dates.sort()
dates
```

```python
def parse_date(datestring):
    return pd.to_datetime(datestring, format="%Y-%m-%d")
```

```python
startdate=parse_date(dates[0])
enddate=parse_date(dates[-1])
print (startdate, ' to ', enddate)
```

```python
index=pd.date_range(startdate, enddate, freq='D')
Graphic1a=pd.DataFrame(index=index, columns=['newcases', 'cases', 'deaths'])
Graphic1a
```

```python
for entry in datalist: 
    date=parse_date(entry['date'])
    for column in ['newcases', 'cases', 'deaths']:
        if pd.isna(Graphic1a.loc[date, column]): 
            value= float(entry[column]) if entry[column]!=None else 0.0
            Graphic1a.loc[date, column]=value            
Graphic1a.fillna(0.0, inplace=True)
            
Graphic1a
```

```python
Graphic1a.plot()
```

```python
Graphic1a.plot(logy=True)
```

```python
with open("data2", "rt") as INFILE:
    Graphic2=json.load(INFILE)
```

```python
Graphic2 #admissions, admissionsage
```

```python
datalist2=Graphic2['data']
datalist2
```

```python
datadic=Graphic2['data'][0]
age=datadic['admissionsage']
datadic
```

```python
ageranges=[x['age'] for x in age] 
print(ageranges)
```

```python
def min_age(agerange):
    agerange=agerange.replace('+','') # remove the + from 90+
    start=agerange.split('_')[0]
    return int(start)
ageranges.sort(key=min_age)
print (ageranges)
```

```python
Graphic2a=pd.DataFrame(index=ageranges, columns=['Admissions'])
for entry in age: # each entry is a dictionary
    ageband=entry['age'] # our index position
    Graphic2a.loc[ageband, 'Admissions']=entry['value']
Graphic2a
```

```python
Graphic2a.plot(kind='bar', y='Admissions')
```
